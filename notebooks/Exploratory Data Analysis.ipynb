```python
data_stats = pd.DataFrame(columns=['feature', 
                                   'min_val', 
                                   'max_val', 
                                   'mean_val', 
                                   'median_val', 
                                   'mean_median_diff',
                                   'std_val',
                                   'norm_min',
                                   'norm_max',
                                   'max_abs_z_value'
                                  ])
ignore_columns = set(['objid', 'run', 'rerun', 'camcol',
                      'field', 'obj', 'ra', 'dec', 'clean',
                      'debiased_elliptical', 'debiased_elliptical_ind',
                      'debiased_spiral', 'debiased_spiral_ind', 'specObjID', 'votes', 'combined_spiral',
                      'spiralanticlock', 'spiralclock', 'debiased_uncertain_ind', 'merger', 'galaxy_type',
                      'elliptical', 'edgeon'
                     ])

# error_columns = data[data.feature.str.find('Err') >= 0]['feature'].values

all_columns = set(data.columns.values)
feature_columns = sorted(list(all_columns - ignore_columns))

for feature in feature_columns:
    if (feature.find('Err') >= 0):
        continue

    d = data[feature]
    d_max = d.max()
    d_min = d.min()
    d_mean = d.mean()
    d_median = d.median()
    d_mean_median_diff = d_mean - d_median
    d_std = d.std()
    d_norm_min = (d_min - d_mean)/d_std
    d_norm_max = (d_max - d_mean)/d_std
    d_max_abs_z_value = max(abs(d_norm_min), abs(d_norm_max))
    
#     feature_stats.iloc[feature] = [d_min, d_max, d_mean, d_median]
    data_stats = data_stats.append({
        'feature': feature,
        'min_val': d_min,
        'max_val': d_max,
        'mean_val': d_mean,
        'median_val': d_median,
        'mean_median_diff': d_mean_median_diff,
        'std_val': d_std,
        'norm_min': d_norm_min,
        'norm_max': d_norm_max,
        'max_abs_z_value': d_max_abs_z_value,
    }, ignore_index=True)

#data_stats
```

### Notes from data stats

* 1237654605862731798 has `zErr` = -1. This is the only record where `zErr` < 0. Will ignore this record.
* 1237665101678117017 has `expRad_u` = 0.0 and 1237662698105995481 has `deVRad_u` = 0.0 both have respective errors as zero. At this point in time will filter out the records rather than trying to find appropriate values.
* There are 96 rows that have `photoz` as -9999. These will need to be removed as there is something wrong with calculation of photoz which might mean that there is something wrong with the photometric data.
* `petroRad_i` and other radii are best suited using the `log(x)` as they are right skewed with a very long tail. Do a pot of `log(x)` has made the values look more normally distributed
* `petro_R90_R50_ratio_u` and other Petrosian radii ratios are best suited using `log(x)`. Astronomers use `log(R90/R50)` as the concentration index.
* `r_i_colour_index` and other colour indexes are best suited using `exp(x)`. The colour index is the difference in magnitudes which are a `log` scale. Using the `exp` is effectively measuring the ratios of fluxes.
* The Mean Stokes parameters are best suited by using `arctan(x)` or `tanh(x)`. The exact meaning for this is not clear but since they were concentrated around being zero but had outliers in the order of 1000s these functions are able to scale the data in.
