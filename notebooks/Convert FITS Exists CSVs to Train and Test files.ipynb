```python
from __future__ import print_function

import numpy as np
import pandas as pd

import sdss_gz_data as sgd

from astropy.io import fits
```

```python
orig_data = sgd.load_data('data/astromonical_data.csv.gz')
```

```python
reindex_data = orig_data.copy()
reindex_data['idx'] = reindex_data['objid']
reindex_data.set_index('idx', inplace=True)
```

```python
train_exists = pd.read_csv('train_exists.csv.bz2')
train_exists.objid = train_exists.objid.astype(str)
test_exists = pd.read_csv('test_exists.csv.bz2')
test_exists.objid = test_exists.objid.astype(str)
```

```python
train_df = pd.DataFrame(columns=orig_data.columns)
for idx, record in enumerate(train_exists.itertuples()):
    if (idx % 1000 == 0):
        print(f'Processed {idx}')

    data_row = reindex_data.loc[record.objid]
    train_df = train_df.append(data_row, ignore_index=True)

train_df
```

```python
test_df = pd.DataFrame(columns=orig_data.columns)
for record in test_exists.itertuples():
    if (idx % 1000 == 0):
        print(f'Processed {idx}')

    data_row = reindex_data.loc[record.objid]
    test_df = test_df.append(data_row, ignore_index=True)

test_df
```

```python
len(train_df), len(test_df)
```

```python
train_df.to_csv('train_objs.csv.bz2', index=False)
test_df.to_csv('test_objs.csv.bz2', index=False)
```

```python
def split_train_and_validation(split_size=0.2):
    from sdss_gz_data import binnify

    indexes = np.arange(len(train_df))
    np.random.shuffle(indexes)

    end_idx = int(np.round(len(train_df) * split_size))
    validate_index = indexes[0:end_idx]
    train_index = indexes[end_idx:-1]

    X_t, X_v = train_df.iloc[train_index], train_df.iloc[validate_index]
    y_t, y_v = train_df.iloc[train_index][['debiased_elliptical_ind', 'z']], train_df.iloc[validate_index][['debiased_elliptical_ind', 'z']]
    
    return X_t, y_t, X_v, y_v

X_train, y_train, X_validate, y_validate = split_train_and_validation()
```
